```python


from nltk import pos_tag,RegexpParser
from nltk.tokenize import word_tokenize
```

```python
data = """
Data analysis is the process of inspecting, cleansing, transforming, and modeling data
"""
```

```python
grammer = "NP:{<DT><JJ><NN>}"
chunker = RegexpParser(grammer)
tags = pos_tag(word_tokenize(data))
```

```python
tree = chunker.parse(tags)
tree.draw()
```

```python
grammer_vb = "VP:{<VBG><NN>}"
chunker2 = RegexpParser(grammer_vb)
tree2 = chunker2.parse(tags)
tree2.draw()
```

Full/Deep Parsing

```python
import spacy
```

```python
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

data = """
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing 
conclusions, and supporting decision-making. Data analysis has multiple facets and approaches, encompassing diverse techniques under a
variety of names, and is used in different business, science, and social science domains.
In today's business world, data analysis plays a role in making decisions more scientific and helping businesses operate more effectively
"""

# Process the text using english pipeline
doc = nlp(data)

# Print dependency relations
for token in doc:
    print(f"{token.text:<15} {token.dep_:<10} {token.head.text:<15} {token.pos_}")

# Visualize one sentence as dependency tree (in Jupyter)
displacy.render(list(doc.sents)[0], style="dep", jupyter=True, options={"distance": 120})
```
